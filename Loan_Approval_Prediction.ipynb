```python
import pandas as pd
df = pd.read_csv('/content/credit.csv')
df.columns
```

```python
x = df[['age', 'income', 'years_at_job', 'credit_score',
       'existing_credit_cards']]
y = df[['approved']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
lrn = LogisticRegression()
lrn.fit(x_train,y_train)
y_pred = lrn.predict(x_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
#Accuracy is 0.72
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(x_train,y_train)
y_pred1 = dtc.predict(x_test)
print(classification_report(y_test,y_pred1))
#Accuracy is 0.94
```

```python
#Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(x_train,y_train)
y_pred2 = rfc.predict(x_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred2))
#Accuracy is 0.95
```

```python
#Gradient Boosting Classifier
from sklearn.ensemble import GradientBoostingClassifier
gbc = GradientBoostingClassifier()
gbc.fit(x_train,y_train)
y_pred3 = gbc.predict(x_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred3))
#Accuracy is 0.95
```

```python
#AdaBoost CLassifier
from sklearn.ensemble import AdaBoostClassifier
abc = AdaBoostClassifier()
abc.fit(x_train,y_train)
y_pred4= abc.predict(x_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred4))
#Accuracy is 0.95
```

```python
#KNeighbors Classifier
from sklearn.neighbors import KNeighborsClassifier
knc = KNeighborsClassifier()
knc.fit(x_test,y_test)
y_pred5 = knc.predict(x_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred5))
#Accuracy is 0.84
```

```python
#Support Vector Classifier
from sklearn.svm import SVC
svc = SVC()
svc.fit(x_test,y_test)
y_pred6 = svc.predict(x_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred6))
#Accuracy is 0.73
```

```python
#Naive Bayes Classifier
from sklearn.naive_bayes import GaussianNB
gbc = GaussianNB()
gbc.fit(x_test,y_test)
y_pred7 = gbc.predict(x_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred7))
#Accuracy is 0.88
```

```python
#Conclusion- Random Forest Classifier, AdaBoost Classifier and Gradient Boosting Classier are the best model with 0.95 accuracy
```
